```python
import preprocess.pre as pre
import pandas as pd
```

```python
df = pd.read_csv('../data/cv.csv')
for index, row in df.iterrows():
    title = str(row['Title'])
    tags = str(row['Tags'])
    body = str(row['Body'])
    title = pre.processbody(title)
    print(title)
    body = pre.processbody(body)
    tags = pre.preprocesstag(tags)
    df.at[index, 'Title'] = title
    df.at[index, 'Tags'] = tags
    df.at[index, 'Body'] = body
df.to_csv('../preprocess/cv-process.csv')
```
